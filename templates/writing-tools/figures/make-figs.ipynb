# Plot of the LJ 12-6 potential function:

```python
def lj_potential(r, sigma=1, epsilon=1, m=12, n=6):
    v = 4*epsilon*((sigma/r)**m-(sigma/r)**n)
    return v

x = np.arange(0.01, 10, 0.01)

fig = plt.figure()

ymax = 3
ymin = -1.5
xmax = 4.0

# Plot the potential
plt.plot(
    x,
    lj_potential(x),
    #label="V(r)"
)

# Plot the force
#plt.plot(x, -np.gradient(lj_potential(x), 0.01), color="maroon", label="F(r)")

# Plot dashed lines showing where V = 0 (r=sigma)
#plt.hlines(y=0, xmin=0, xmax=1, linestyles="--", colors="k")
#plt.vlines(x=1, ymin=ymin, ymax=0, linestyles="--", colors="k")
#plt.plot(1, 0, marker="o", c="k", markersize=3, label="$r = 1\sigma$", linewidth=0)

# Plot dashed lines showing where V = min (r=1.12sigma)
plt.hlines(y=-1, xmin=0, xmax=2**(1/6), linestyles="--", colors="g")
plt.vlines(x=2**(1/6), ymin=ymin, ymax=-1, linestyles="--", colors="g")
plt.plot(1.12, -1, marker="o", c="g", markersize=3, label="$r_{min} (r = 2^{1/6}\sigma)$", linewidth=0)
plt.text(1.16, -1.15, "$r_{min}$")

#plt.title("12-6 Lennard Jones Potential")
plt.xlabel("$r\,(\sigma)$")
plt.ylabel("$V(r)\,(\epsilon)$")
plt.ylim(ymin, ymax)
plt.xlim(0, xmax)
#plt.legend()

# Save the fig
plt.savefig("lj-potential-no-title.png")
```
